```python
import matplotlib.pyplot as plt
import numpy as np

T = [30]
horas = 0.01
k = 0.4056
t_muerte=12
t_final = t_muerte+horas

while T[-1] < 37:
    T.append(horas*k*(T[-1]-27)+T[-1])
    t_final-=horas

plt.plot(T)
print(t_final)
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = [1]
h = 0.01
to = 0
k = 0.0009906
T_final=to+h

while T_final <12:
    x.append(h*k*x[-1]*(1000-x[-1])+x[-1])
    T_final+=h

print(x)
    

plt.subplot(121)
plt.plot(np.arange(to,12+h,h),x)
plt.subplot(122)
plt.plot(np.arange(to,12+h,h) , 1000 - np.array(x))
plt.grid()
```
