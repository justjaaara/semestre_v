```python
#Sin
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2*np.pi, 0.1)
x1 = np.linspace(0,2*np.pi, 1000)
plt.plot(x1, np.sin(x1), 'r')
```

```python
#Cos
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2*np.pi, 0.1)
x1 = np.linspace(0,2*np.pi, 1000)
plt.plot(x1, np.cos(x1), 'r')
```
